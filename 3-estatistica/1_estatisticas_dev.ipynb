```python
import pandas as pd
```

```python
pd.__version__
```

```python
dict_medidas = {
    'idade': [15,18,25,25,40,55,58,60,80],
    'altura': [160,162,165,168,172,174,174,174,176]
}
```

```python
dict_medidas
```

```python
df_medidas = pd.DataFrame.from_dict(dict_medidas)
```

```python
df_medidas
```

## Medidas Resumo

### Medidas de Posição

```python
# média
df_medidas["idade"].mean()
```

```python
# mediana
df_medidas["idade"].median()
```

```python
# moda
df_medidas["idade"].mode()
```

```python
# moda
df_medidas["altura"].mode()
```

### Medidas de Dispersão

```python
# variância
df_medidas.idade.var()
```

```python
# desvio padrão
df_medidas.idade.std()
```

```python
# coeficiente de variação
df_medidas.idade.std() / df_medidas.idade.mean() * 100
```

```python
# coeficiência de variação da altura
df_medidas.altura.std() / df_medidas.altura.mean() * 100
```

### Medidas de Forma

```python
# Assimetria - Idade
df_medidas.idade.skew()
```

```python
# Assimetria - Altura
df_medidas.altura.skew()
```

```python
# Curtose - Idade
df_medidas.idade.kurtosis()
```

```python
# Curtose - Altura
df_medidas.altura.kurtosis()
```

```python
# Obter medidas estatísticas 
df_medidas.idade.describe()
```

### Correlação de variáveis

```python
# Correlação do DataFrame, segundo Pearson
df_medidas.corr(method='pearson')
```

```python
# Correlação entre duas variáveis
df_medidas.idade.corr(df_medidas.altura)
```

```python
# Correlação do DataFrame, segundo Spearman
df_medidas.corr(method='spearman')
```

## Representações gráficas

```python
# Histograma - Idade
df_medidas.idade.hist()
```

```python
# Histograma - Altura
df_medidas.altura.hist()
```

### Criar DataFrame de Vendas

```python
dict_vendas = {
    "categoria": ["masculino", "feminino", "infantil", "casa"],
    "valor": [400000, 700000, 250000, 580000],
    "quantidade": [3000,5000,1500,2500]
}
```

```python
dict_vendas
```

```python
df_vendas = pd.DataFrame.from_dict(dict_vendas)
```

```python
df_vendas
```

```python
# Gráfico de Barras Vertical - Valor
df_vendas.plot.bar(x='categoria', y='valor')
```

```python
# Gráfico de barras vertical - Valor Ordenado
df_vendas.sort_values(by='valor', ascending=False).plot.bar(x='categoria', y='valor') 
```

```python
# Gráfico de barras horizontal - Quantidade
df_vendas.plot.barh(x='categoria', y='quantidade')
```

```python
# Gráfico de barras horizontal - Quantidade (Ordenado - Ranking)
df_vendas.sort_values(by='quantidade').plot.barh(x='categoria', y='quantidade')
```

```python
# Gráfico de Dispersão
df_medidas.plot.scatter(x='idade', y='altura')
```

```python
# Gráfico de boxplot (diagrama de caixa) - Idade
df_medidas.idade.plot.box()
```

```python
# Gráfico de boxplot - Altura
df_medidas.altura.plot.box()
```
