# Análise Bivariada

```python
# Hipótese - Clientes com contrato do tipo mensal são mais propensos ao Churn
# Construir uma tabela de contingência entre tipo de contrato e churn - Quantidade
pd.crosstab(df_churn['Churn'], df_churn['Contract'])
```

```python
# Adicionando coluna de somatório
pd.crosstab(df_churn['Churn'], df_churn['Contract'], margins=True, margins_name='Total')
```

```python
# Construir uma tabela de contingência entre tipo de contrato e churn - Proporção
pd.crosstab(df_churn['Churn'], df_churn['Contract'], normalize='index', margins=True, margins_name='Total')
```

Considerando o público total, os clientes de contrato mensal representam 55%. Porém, se considerarmos apenas os clientes que abandonaram o serviço, 88% tinham contrato mensal. Desta forma podemos afirmar que há uma correlação entre o tipo de contrato e o abandono.

```python
# Avaliar a correção entre duas variáveis categóricas (qualitativas nominais)

# Executar um teste de hipótese chamado `Chi-Square` ou `Qui-Quadrado` (também criado por Pearson)

# Num teste de hipótese, duas hipóteses são formuladas:
# H0 (Hipótese Nula): as duas variáveis são independentes
# H1 (Hipótese Complementar): as duas variáveis são dependentes

# O teste serve para confirmar ou recusar a hipótese nula (H0)
# Quanto a probabilidade de observarmos que H0 é inferior a 0.05 (p-value), recusamos a hipótese e seguimos com a complementar
```

```python
# Criar um dataframe da crosstable (sem totais)
df_crosstab_churn_contract = pd.crosstab(df_churn['Churn'], df_churn['Contract'])
```

```python
# Calcular os Scores e P-Values
from scipy.stats import chi2_contingency
```

```python
# Tupla = Lista imutável
chi_scores_churn_contract = chi2_contingency(df_crosstab_churn_contract)
```

```python
chi_scores_churn_contract
```

```python
scores_churn_contract = pd.Series(chi_scores_churn_contract[0])
```

```python
pvalues_churn_contract = pd.Series(chi_scores_churn_contract[1])
```

```python
# Apresentar números com decimais sem a notação científica
pd.set_option('display.float_format', lambda x: '%.15f' % x)
```

```python
# Criar dataframe com os resultados
df_chi_scores_churn_contract = pd.DataFrame({'Qui2': scores_churn_contract, 'p-value': pvalues_churn_contract})
```

```python
# Analisar Scores e P-Values
df_chi_scores_churn_contract
```

Conforme o teste estatístico (Chi-Square), o P-Value <= 0.05, desta forma rejeitamos a hipótese nula, ou seja, as variáveis não são independentes. Pelo Qui2 alto, podemos afirmar que há uma forte correlação.

```python
# Avaliar a correlação entre uma variável qualitativa e quantitativa
# Usar o mesmo teste de Chi-Square, mas transformar a variável quantitativa em qualitativa

# Hipótese: Cliente com menos de 6 meses de contrato é mais propenso ao Churn

# Criar uma coluna nova no dataframe com base numa condição de outra coluna
# Coluna = "TenureLowerThan6Months"

import numpy as np
```

```python
# Criar coluna nova
df_churn['TenureLowerThan6Months'] = np.where(df_churn['tenure'] < 6, 'Yes', 'No')
```

```python
# Checar o dataframe
df_churn
```

```python
# Criar um dataframe da crosstable (sem totais)
df_crosstab_churn_tenure = pd.crosstab(df_churn['Churn'], df_churn['TenureLowerThan6Months'])
```

```python
# Calcular os scores
chi_scores_churn_tenure = chi2_contingency(df_crosstab_churn_tenure)
```

```python
scores_churn_tenure = pd.Series(chi_scores_churn_tenure[0])
pvalues_churn_tenure = pd.Series(chi_scores_churn_tenure[1])
```

```python
# Criar um dataframe com os resultados
df_chi_scores_churn_tenure = pd.DataFrame({'Qui2': scores_churn_tenure, 'p-value': pvalues_churn_tenure})
```

```python
df_chi_scores_churn_tenure
```

Como o P-Value <= 0.05, rejeitamos a H0, ou seja, as variáveis não são independentes. E vale mencionar que com base do Qui2 resultante das 2 análises entre Churn e Tempo de Contrato < 6 meses é menor do que a correlação entre Churn e Tipo de Contrato.

```python
# Correlação entre variáveis numéricas

# Tenure - TotalCharges - A intuição é que quanto mais tempo de contrato maior o valor pago.

# Correlação entre 2 variáveis numéricas - Pearson
df_churn['tenure'].corr(df_churn['TotalCharges'])
```

```python
# Correlação entre 2 variáveis numéricas - Spearman
df_churn['tenure'].corr(df_churn['TotalCharges'], method='spearman')
```

A uma forte correlação entre `Tenure` e `TotalCharges`, usando métodos estatísticos de correlação.

```python
# Criar um novo dataframe somente com os clientes (month-to-month)
df_churn_month = df_churn[df_churn['Contract'] == 'Month-to-month']
```

```python
df_churn_month
```

```python
# Histograma - Checar visualmente se os dados seguem uma distribuição normal
df_churn_month['TotalCharges'].plot.hist()
```

```python
# Método de Tukey - IQR (distribuição não normal)
# IQR - Range Interquartil
# IQR = Q3 - Q1

total_charges_Q1 = df_churn_month['TotalCharges'].quantile(0.25)
total_charges_Q3 = df_churn_month['TotalCharges'].quantile(0.75)
iqr_total_charges = total_charges_Q3 - total_charges_Q1

iqr_total_charges
```

```python
# Limites - inferior e superior
inf_limit = total_charges_Q1 - (iqr_total_charges * 1.5)
sup_limit = total_charges_Q3 + (iqr_total_charges * 1.5)

inf_limit, sup_limit
```

```python
df_churn_month[(df_churn_month['TotalCharges'] < inf_limit) | (df_churn_month['TotalCharges'] > sup_limit)]
```

```python
# Método - Zscore (distribuição normal)
# Indica quantos desvios padrões um ponto específico de dados está distante da média

# xscore = (x - média) / desvio padrão

from scipy.stats import zscore
```

```python
z = np.abs(zscore(df_churn_month['TotalCharges']))
```

```python
z
```

```python
df_churn_month[z > 3.0]
```
