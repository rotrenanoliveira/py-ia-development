# Transformação de Dados

```python
# Transformar coluna TotalCharges de String para Float - Abordagem 1 com astype
# Essa abordagem só funciona se o tipo de dado da coluna for um campo válido para o que você precisa mas não está no formato esperado
# df_contracts['TotalCharges'] = df_contracts['TotalCharges'].astype(float)
```

```python
# Transformar coluna TotalCharges de String para Float - Abordagem 2 (to_numeric)
# Essa abordagem só funciona se o tipo de dado da coluna for um campo válido para o que você precisa mas não está no formato esperado
# df_contracts['TotalCharges'] = pd.to_numeric(df_contracts['TotalCharges'])
```

```python
# Transformar coluna TotalCharges de String para Float - Abordagem 3 (to_numeric com coerce)
df_contracts['TotalCharges'] = pd.to_numeric(df_contracts['TotalCharges'], errors='coerce')
```

```python
# Validando a transformação
df_contracts.info()
```

Após a transformação da coluna TotalCharges do DataFrame Contracts, a coluna passou a ter 11 valores ausentes.

# Unificar DataFrames de Customers, Services e Contracts

```python
# Contar o número de registros por dataframe
len(df_customers)
```

```python
len(df_contracts)
```

```python
len(df_services)
```

```python
# Renomear colunas
# df_services = df_services.rename(columns={'customerID': 'customer_id'}, inplace=True)
```

```python
# Unificar dataframe de Customers e Services, temporário
df_temp = df_customers.merge(df_services, on='customerID')
```

```python
df_temp.info()
```

```python
# Unificando o dataframe temporário com Contracts, usando colunas de junção com nomes distintos
df_churn_temp = df_temp.merge(df_contracts, left_on=['customerID'], right_on=['customerID'])
```

```python
df_churn_temp.info()
```

```python
# Unificar os três dataframe ao mesmo tempo, usando nomes de colunas diferentes
df_churn = df_customers.merge(df_services, on=['customerID']).merge(df_contracts, left_on=['customerID'], right_on=['customerID'])
```
